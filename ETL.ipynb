El mismo consta de dos datasets con los siguientes datos

El primero movies_dataset.csv tiene las siguientes características:
- budget: El presupuesto con el que se realizó la película.
- genre: El género de la película, acción, comedia, thriller, etc.
- homepage: Un enlace a la página principal de la película.
- id: En realidad, es el movie_id tal como está en el primer conjunto de datos.
- keywords: Las palabras clave o etiquetas relacionadas con la película.
- original_language: El idioma en el que se realizó la película.
- original_title: El título de la película antes de la traducción o adaptación.
- overview: Una breve descripción de la película.
- popularity: Una cantidad numérica que especifica la popularidad de la película.
- production_companies: La productora de la película.
- production_countries: El país en el que se produjo.
- release_date: La fecha en la que se estrenó.
- revenue: Los ingresos generados por la película en todo el mundo.
- runtime: La duración de la película en minutos.
- status: "Estrenada" o "Rumor".
- tagline: Eslogan de la película.
- title: Título de la película.
- vote_average: Calificación promedio que recibió la película.
- vote_count: La cantidad de votos recibidos.

El segundo Credits.csv contiene las siguientes características:
- movie_id: Un identificador único para cada película.
- cast: El nombre de los actores principales y de reparto.
- crew: El nombre del director, editor, compositor, escritor, etc.

Voy a unir los dos conjuntos de datos en la columna 'id'.

```python
# Carga de los dataset
df_movies = pd.read_csv('movies_dataset.csv')
df_credits = pd.read_csv('credits.csv')
```

```python
# Para poder unir ambos dataframe debo pasar el id de movies al mismo tipo que el id de credits
# Si bien id de credits es un int lo paso a formato str para que no me de error al utilizar el merge
df_movies['id'] = df_movies['id'].astype(str)
df_credits['id'] = df_credits['id'].astype(str)
```

```python
# Voy a combinar ambos dataframe
df_movies = df_movies.merge(df_credits, on='id')
```

```python
# Veo como quedo el df unido
df_movies.head(5)
```

```python
# Voy a eliminar las columnas video, imdb_id, adult, original_title, poster_path y homepage que no serán utilizadas 
df_movies = df_movies.drop('video', axis=1)
df_movies = df_movies.drop('imdb_id', axis=1)
df_movies = df_movies.drop('adult', axis=1)
df_movies= df_movies.drop('original_title', axis=1)
df_movies = df_movies.drop('poster_path', axis=1)
df_movies = df_movies.drop('homepage', axis=1)
```

```python
# El campo Runtime repesenta los minutos que dura una pelicula, actualmente el df lo tomo como float con lo cual, se va a tranformar
df_movies['runtime'] = pd.to_numeric(df_movies['runtime'], errors='coerce').astype(float)
df_movies['runtime'] = df_movies['runtime'].fillna(0).astype(int)
```

```python
#Relleno los valores NaN de revenue  y budget por O (cero)
df_movies['revenue'] = df_movies['revenue'].fillna(0)
df_movies['budget'] = df_movies['budget'].fillna(0)
```

```python
# Voy a eliminar filas con valores no validos en la columna release_date
df_movies = df_movies.dropna(subset=['release_date'])
# Relleno los valores no validos con un valor por defecto 
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], format='%Y-%m-%d', errors='coerce').fillna(pd.NaT)
# Eliminamos los valores que rellenamos en el codigo de arriba
df_movies.dropna(subset=['release_date'], inplace=True)
```

```python
# Creo la columna de release_year con el año de release
df_movies.loc[:, 'release_year'] = df_movies['release_date'].dt.year
```

```python
# Convierto la columna revenue y budget a valores numeroricos para poder usarlos en la creacion de la columna retorno de inversion 
df_movies['revenue'] = pd.to_numeric(df_movies['revenue'], errors='coerce') 
df_movies['budget'] = pd.to_numeric(df_movies['budget'], errors='coerce')

# Voy a crear la columna return (retorno de inversion)
df_movies['return'] = df_movies['revenue'] / df_movies['budget']
df_movies['return'].fillna(0, inplace=True)
```

```python
df_movies.head()
```

```python
# Haciendo un chequeo visual veo que en return figura una valor en inf  con lo cual lo convierto a 0 como debe figurar
df_movies['return'] = df_movies['return'].replace(np.inf, 0)
```

```python
# Voy a generar una columna con el dia para una de las funciones que me estan solicitando 
df_movies['release_day'] = df_movies['release_date'].dt.day_name(locale='es_ES')
```

```python
# Voy a generar una columna con el mes para una de las funciones que me estan solicitando 
df_movies['release_month'] = df_movies['release_date'].dt.month_name(locale='es_ES')
```

```python
df_movies
```

```python
# Creo un df_mes y dia para la funciones solicitada
df_mes = df_movies[['id', 'release_month']]
df_dia = df_movies[['id', 'release_day']]
df_movies_titulo = df_movies[['id','title','vote_average','vote_count', 'release_year']]
df_movies_score = df_movies[['id','title','release_year','popularity']]
```

```python
# Se exportan los primeros 4 dataset, elegi el formato parquet ya que es el mas liviano y performante.
df_mes.to_parquet('df_mes', engine='fastparquet')
df_dia.to_parquet('df_dia', engine='fastparquet')
df_movies_titulo.to_parquet('df_movies_titulo', engine='fastparquet')
df_movies_score.to_parquet('df_movies_score', engine='fastparquet')
```

```python
# Voy a trabajar sobre la columna genres primero la transformamos en un lista con la appy literal_eval donde cada valor se la columna se transforma en una lista de python
# Luego aplico una lambda para generar una lista con los nombres, en caso de estar vacio se genera una lista vacia.
df_movies['genres'] = df_movies['genres'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
# Cheque como quedo genres 
df_movies.head(5)
```

```python
# Creo una funcion para obtener los nombres de los Directores, esto no va a servir para los actores
# ya que tendra que ser en un formato de lista porque son varios por peliculas

def bus_director(x):
    '''
    Esta funcion busca en una lista por diccionario que tiene la clave 'job' con el valor de director, una vez 
    se encuentra el diccionario de director se devuelve el valor asociado a la clave 'name'.
    De no encontrar valor se devolvera un valor nulo.
    '''
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# Convierto la columna 'crew' en una lista de diccionarios
df_movies['crew'] = df_movies['crew'].apply(literal_eval)
```

```python
df_movies['director'] = df_movies['crew'].apply(bus_director)
```

```python
df_movies.head()
```

```python
# Voy a generar una lista con los nombres de los actores y guardarlo en una nueva columna que se llame actores 
# Utilizo la funcion literal_eval para convertirla en una lista propia de python 
# x itera con cada elemento de i en la lista y extrae el nombre del actor usando la clave name, 
# luego devuelve una lista con los nombres de los actores 
df_movies['actores'] = df_movies['cast'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
# Chequeo la columna actores 

df_movies.head(10)
```

```python
# Voy a utilizar la misma linea de codigo que utilice para genres y cast en production_companies, production_countries 
df_movies['production_companies'] = df_movies['production_companies'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df_movies['production_countries'] = df_movies['production_countries'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
# da error, se utiliza el modulo ast que da  herramientas para analizar y manipular árboles sintácticos abstractos, con eso deja de dar error y nos permite correr el lambda
df_movies['belongs_to_collection'] = df_movies['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if isinstance(x, str) else np.nan)
```

```python
# Chequeo los resultados en production_companies y production_countries
df_movies.head()
```

```python
# Por fines practicos voy a generar un df sin los datos anidados y en formato parquet por si en algun momento lo necesito 
# Elimino cast y crew ya que tienen datos anidados
df_movies_final = df_movies.drop('cast', axis=1, inplace=True)
df_movies_final = df_movies.drop('crew', axis=1)
```

```python
# Exporto df_movies_final en parquet 
df_movies_final.to_parquet('df_movies_final', engine='fastparquet')
```
